```python
departamentosmapa = departamentosmapa.replace(["05", "08"], [5,8])
departamentosmapa["DPTO_CCDGO"] = pd.to_numeric(departamentosmapa["DPTO_CCDGO"])
departamentosmapa
```

```python
mapadf = pd.merge (Departamentos, departamentosmapa, on= ["DPTO_CCDGO"])
mapadf
```

```python
dptosmapa = geopandas.GeoDataFrame(mapadf, geometry='geometry')
```

```python
ax3 = plt.axes()
dptosmapa.plot(
    ax=ax3,
    column="%informalidad",
    legend=True,
    cmap="GnBu")
plt.title("Informalidad por departamentos")
plt.savefig("Mapadptos.jpg")
```

```python
colors = sns.color_palette('Blues')[0:2]

fig, ax = plt.subplots(figsize=(3,3), dpi=100)

ax.bar(Etnias["Etnia"], Etnias["%informalidad"], color=colors)
ax.set_title("Informalidad laboral por etnia (%)")
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.savefig("Etniasgrafico.jpg")
plt.show()
```

```python
colores = {"Mujer":"lightpink", "Hombre":"lightblue"}
colorsexo = Sexo['Sexo'].map(colores)
fig, ax = plt.subplots(figsize=(4,1), dpi=100)
ax.barh(Sexo["Sexo"], width = Sexo["%Informalidad"], color = colorsexo)
ax.set_title("Informalidad laboral por sexo", fontsize=6)
plt.savefig("Sexos.jpg")
plt.show()
```

```python
colores = {"Mujer":"lightpink", "Hombre":"lightblue"}
colorsexo = Sexo['Sexo'].map(colores)

#create pie chart
plt.figure(figsize=(4, 4), dpi=100)
plt.pie(Sexo["%Informalidad"], colors = colorsexo, autopct='%.0f%%')
plt.legend(labels=Sexo["Sexo"])
plt.title("Proporción informalidad laboral por sexo", fontsize=6)
plt.savefig("ProporcionSexo.jpg")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4,3), dpi=100)
ax.stem(Educacion["Nivel"], Educacion["%Informalidad"], label= Educacion["Nivel"], linefmt = "grey", markerfmt = "oc", basefmt = 'c:')
ax.set_title("Informalidad por nivel educativo (%)*")
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
plt.savefig("Educacion.jpg")
plt.show()
```
